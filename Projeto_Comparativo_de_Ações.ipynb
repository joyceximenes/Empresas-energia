```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dados = pd.read_excel('/content/drive/MyDrive/Study analise de dados/Data Viking/Projeto Empresas de Energia/Dados_Empresas_Energia.xlsx')
dados.head()
```

```python
dados.set_index('Data', inplace=True)
```

**plotando gráfico**

```python
plt.style.use('seaborn-darkgrid')
plt.figure(figsize=(12,6))
plt.title('Analise de ações das empresas', loc='center')

# plot da petrobras
plt.plot(dados.index, dados['Petrobras'], color='#008c4a', linewidth = 4) # se organiza pelas datas, informa a coluna, determina cor, alarga linha
plt.text(dados.index[-1], dados['Petrobras'].tail(1), 'Petrobras', color='#008c4a', size='large', horizontalalignment='right') 
  # incluir informação no gráfico, acessa o ultimo valor e escreve isso lá (tanto no index como na coluna)

# Plot de todas as colunas
for Coluna in dados.columns[1:]: # nao inclui a petrobras, mostra todas as outras
  # Plot das outras ações
  plt.plot( dados.index, dados[Coluna], color='gray', linewidth=2, alpha=0.7 )
  # Texto das outras ações
  plt.text( dados.index[-1], dados[Coluna].tail(1), Coluna, color='gray', size='large', horizontalalignment='left' )

# Labels
plt.xlabel('Período')
plt.ylabel('Preço de Fechamento (R$)');
```
